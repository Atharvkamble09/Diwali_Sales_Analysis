# Exploratory Data Analysis

# Gender

```python
diwali_df.columns
```

```python
# To show the graph
sns.set(rc ={'figure.figsize':(4,2)})#Used to change the figure size

ax = sns.countplot(x='Gender',data=diwali_df) 

for bars in ax.containers:
    ax.bar_label(bars);
```

```python
# Here firs we group the values with gender
# In second we add the values in the amount
# In third we sort the amount values by descending

diwali_df.groupby(['Gender'],as_index =False)['Amount'].sum().sort_values(by ='Amount',ascending=False)
```

```python
shooping = diwali_df.groupby(['Gender'],as_index =False)['Amount'].sum().sort_values(by ='Amount',ascending=False)
sns.barplot(x='Gender', y='Amount',data=shooping);
```

From the above we can see that most of the buyers are females and even the purchasing power of the female is also greater than the men

# Age

```python
ax = sns.countplot(data = diwali_df , x='Age Group',hue = 'Gender') 

for bars in ax.containers:
    ax.bar_label(bars);
```

```python
shooping = diwali_df.groupby(['Age Group'], as_index =False)['Amount'].sum().sort_values(by ='Amount',ascending=False)
sns.barplot(x='Age Group', y='Amount',data=shooping);
```

From above we can see that most of the buyes are in age group 26-35

# state

```python
# Total number of orders from top 10 states

sales_state =diwali_df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by ='Orders',ascending=False).head(10)

sns.set(rc ={'figure.figsize':(16,8)})#Used to change the figure size

sns.barplot(data=sales_state,x='State',y='Orders');
```

```python
# Total amount/sales from top 20 states

sales_state =diwali_df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending=False).head(20)

sns.set(rc ={'figure.figsize':(28,10)})#Used to change the figure size

sns.barplot(data=sales_state,x='State',y='Amount');
```

From above we can see that most of the orders are from uttar pradesh maharashtra and karnataka similar in case of amount of the products

# Marital_Status

```python
ax = sns.countplot(data = diwali_df , x='Marital_Status') 
sns.set(rc ={'figure.figsize':(4,2)})
for bars in ax.containers:
    ax.bar_label(bars);
```

```python
sales_state =diwali_df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending=False).head(20)

sns.set(rc ={'figure.figsize':(8,4)})

sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue = 'Gender');
```

From above we can see that most of the buyer are married(women) and they also have the high purchesing power

# Occupation

```python
diwali_df.columns
```

```python
sns.set(rc ={'figure.figsize':(20,8)})
ax = sns.countplot(data = diwali_df , x='Occupation') 
for bars in ax.containers:
    ax.bar_label(bars);
```

```python
sales =diwali_df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending=False)

sns.set(rc ={'figure.figsize':(20,8)})

sns.barplot(data=sales,x='Occupation',y='Amount');
```

From above we can see that most of the buyers are from IT sector,Healthcare and Aviation.

# Product_Category

```python
sns.set(rc ={'figure.figsize':(26,8)})
ax = sns.countplot(data = diwali_df , x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars);
```

```python
sales =diwali_df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending=False).head(10)

sns.set(rc ={'figure.figsize':(20,8)})

sns.barplot(data=sales,x='Product_Category',y='Amount');
```

From above we can see that most of the sold products are Food , Cloths and Electronics Gadgets.

```python
sales =diwali_df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by ='Orders',ascending=False).head(10)

sns.set(rc ={'figure.figsize':(20,8)})

sns.barplot(data=sales,x='Product_ID',y='Orders');
```

```python
# Top 10 most sold products
#nlargest(10) is same as head(10)
fig1 , ax1=plt.subplots(figsize=(16,8))
diwali_df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending = False).plot(kind = 'bar');
```

# Conclusion:

Married women Age group 26-35 years from UP , Maharashtra and Karnataka working in IT , Healthcare and Aviation are more likely to buy products from Food , Clothing and Electronics categories.
